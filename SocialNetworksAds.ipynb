# Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

# Importing the dataset

```python
df = pd.read_csv("SocialNetworkAds.csv")
df
```

# Extracing only the independent variables

```python
x = df.iloc[:,[2,3]].values
x
```

# Extracting only the dependent variables

```python
y = df.iloc[:,4].values
y
```

# Visualising the dataset by drawing a correlation map

```python
sns.heatmap(df.corr())
```

# Splitting the dataset into the Traning set and Test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state =0)
```

# Feature scaling

```python
import sklearn
from sklearn.preprocessing import StandardScaler
sc_x =StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

# Fitting Logistics Regression to Training Dataset

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state =0)
classifier.fit(x_train,y_train)
```

# Predicting the Test Set Results

```python
y_pred = classifier.predict(x_test)
y_pred
```

# Confusion Matrix Evaluation

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```
